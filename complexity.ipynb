# Complexity

When you have a piece of code, how do you reason about its performance? How can you decide which algorighm will perform the best as the input size grows?

You can of course measure the wall-clock time when running the code on differnet inputs. If you vary the size of the inputs, you might be able to derive a function that describes the time needed for differnet inputs. Such a function might say that time $t$ (measured in milliseconds) needed to process an input of size $n$ approximately
$$
f(n) = 200 + 10 n + 3 n^2.
$$
Another implementation that solves the same problem might have a running time described as
$$
g(n) = 800 + 270 n.
$$

This is the running time on *your* machine. If I run the same code, I might end up with
$$
f(n) = 300 + 5 n + n^2
$$
and
$$
g(n) = 500 + 200 n.
$$
We get differnet numbers and different functions... Which function is *right* and which function should we use when we talk about the performance of this code? Which function should we expect to see when we put the code in production? What if we have three differnet kinds of machines in our production environment -- how can we avoid having to profile the code on every type of machine?

To get a feeling for complexity of functions, we will look at three implementations of the same function: removing duplicates from a list (ignoring the order). The first implementation is simple:

```python
def remove_duplicates1(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result
```

The second implementation sorts the list first. It then scans through it, only emitting the first element in a sorted group:

```python
def remove_duplicates2(lst):
    last_item = None
    result = []
    for item in sorted(lst):
        if item != last_item:
            result.append(item)
            last_item = item
    return result
```

The third implementation uses a set to remember already seen items:

```python
def remove_duplicates3(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            result.append(item)
            seen.add(item)
    return result
```

We can measure the running time on differently sized inputs using the timeit module and plot this using matplotlib:

```python
import random
funcs = [remove_duplicates1,
         remove_duplicates2,
         remove_duplicates3]
sizes = range(0, 200, 50) + range(200, 2001, 200)
times = [[], [], []]
for n in sizes:
    lst = range(n // 2) + range(n // 4) + range(n // 8) + range(n // 8)
    random.shuffle(lst)
    for i, func in enumerate(funcs):
        result = %timeit -qo func(lst)
        times[i].append(1000 * result.best)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.xlabel("n")
plt.ylabel("ms")
plt.plot(sizes, times[0], 'r--', label='remove_duplicates1')
plt.legend(loc='upper left')

plt.subplot(1, 2, 2)
plt.xlabel("n")
plt.ylabel("ms")
plt.plot(sizes, times[1], 'b--', label='remove_duplicates2')
plt.plot(sizes, times[2], 'g--', label='remove_duplicates3')
plt.legend(loc='upper left')
```

## Asymptotic Complexity

To avoid reasoning about the running time on specific computers, and to avoid letting small terms obscure the picture, algorithms are measured based on their asymptotic performance. Asymptotic performance shows how the algorithm behaves as the input size grows towards infinity. This is interesting since a lot of code is "fast enough" when the input size is small: some problems don't show up when you test with 10 or 100 customers in your database on your local laptop. However, when you have 50,000 customers in the production database, performance problems can show up.

If we zoom in on the graphs above, we see that all three functions behave almost identically for up to 50 elements, and that the two fast functions cannot be distinguished when we only consider 200 elements:

```python
cutoff = 5
#plt.figure(figsize=(7, 5))

plt.xlabel("n")
plt.ylabel("ms")
plt.plot(sizes[:cutoff], times[0][:cutoff], 'r--', label='remove_duplicates1')
plt.plot(sizes[:cutoff], times[1][:cutoff], 'b--', label='remove_duplicates2')
plt.plot(sizes[:cutoff], times[2][:cutoff], 'g--', label='remove_duplicates3')
plt.legend(loc='upper left')
```

```python
for i, n in enumerate(sizes[:cutoff]):
    print '%3d:' % n, ' '.join('%5.1f µs' % (1000 * t[i]) for t in times)
```

The two fast functions are within 1-2 microseconds of each other -- a tiny difference that you won't notice when testing by hand. Even the slow function is very fast on these input sizes and the difference won't be apparent when running tests or unit tests. However, when we go to larger input sizes, the difference becomes larger and larger: on 100 elements, the first function takes twice as long as the fast functions, on 200 elments, it takes three times as long. Even the fast functions differ when the input size becomes sufficiently large, as we saw on the first plots.

This is the difference we seek to capture when we look at asymptotic complexity: we seek to classify `remove_duplicates1`, `remove_duplicates2`, and `remove_duplicates3` into different complexity classes. Concretely, we will say that `remove_duplicates1` has $O(n^2)$ (quadratic) running time, that `remove_duplicates2` has $O(n \log n)$ running time, and that `remove_duplicates3` has $O(n)$ (linear) complexity.

The $O(g(n))$ notation says that the running time is on the *order of* $g$. The definition is that $f \in O(g(n))$ if and only iff:
$$
\exists n_o \exists c \forall n \geq n_0\colon f(n) \leq c g(n)
$$
In other words: a function $f$ is in $O(g(n))$ when $f$ becomes dominated by $c g$ for some constant factor $c$ and from some point $n_o$ onwards. This means that the function $f$ grows at most as fast as $g$ when $n$ becomes large.

Because the relation must only hold for large $n$, the function has some time to "start" before it must be fast. So if
$$
f(n) = 1000 + n
$$
we have $f \in O(n)$ since we can select $n_0 = 500$ and $c = 3$ so that $f(n) = 1000 + n \leq 3 n = c g(n)$ for all $n \geq 500$. There are many choices for $c$ and $n_0$ that satisfies the definition -- we only need to find a single pair.
